### 크롤링할 키워드 한개일 때 

```python
key = input("검색할 단어를 입력하세요 :")
#site = 'https://search.shopping.naver.com/search/all?query=' + key + '&frm=NVSCPRO&bt=0'
site = 'https://search.shopping.naver.com/search/all?query=' + key
```

```python
ad_product_list = []

    
# 셀레늄으로 웹브라우저를 띄우고 해당주소로 이동합니다.
driver = Chrome()
driver.get(site)
driver.implicitly_wait(3)
    
body = driver.find_element_by_tag_name("body")
    
num_of_pagedowns = 20
    
while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    f = uniform(0.3, 1.0)   # 0.2부터 1.0사이의 임의의 float
    time.sleep(f)
    num_of_pagedowns -= 1
        
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')
sp = soup.select('li.basicList_item__2XT81.ad')
    
i = 0

sp

for i in range(8):
    try:
        product = sp[i]
        product_title = product.select_one('a.basicList_link__1MaTN')
        ad_check = product.select_one('button.ad_ad_stk__12U34')
        store_name = product.select_one('a.basicList_mall__sbVax')
        
        ad_product_list.append(
            {
                "num" : i+1,
                "keyword" : key,
                "product" : product_title.text,
                "ad"  : ad_check.text,
                "store" : store_name.text
            }
        )
    except:
        ad_product_list.append(
            {
                "num" : i+1,
                "keyword" : key,
                "product" : " ",
                "ad"  : " ",
                "store" : " "
            }
        )

driver.close()
    
    
ad_data_list = pd.DataFrame(ad_product_list,columns=['num','keyword','product','ad', 'store'])
ad_data_list = ad_data_list.drop_duplicates()
#ad_data_list.to_excel('ad_data_list.xlsx')
ad_data_list
```
